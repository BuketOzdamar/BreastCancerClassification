```python
import numpy as np
import pandas as pd
from numpy import loadtxt

#Verisetini oku;
data_csv = r"D:\\\\\data.csv"
data = pd.read_csv(data_csv)
#print(data)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn import preprocessing

#İyi huylu-kötü huyluları 0 ve 1 olarak çeviriyoruz.
Malignite =  data.iloc[:,10:11].values
#print(Malignite)

Malignite[:,0]=LabelEncoder().fit_transform(Malignite[:,0])
#print(Malignite)

y = Malignite
print(y)

X = data.drop(["Id","Malignite"],axis=1)
#print(X)
```

```python
#Train-test kümesi oluşturma;
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from keras.models import Sequential
dl = Sequential()
```

```python
from keras.layers import Dense, Activation
dl.add(Dense(5,input_dim=9, kernel_initializer="normal", activation="relu"))
dl.add(Dense(5, kernel_initializer="normal", activation="relu"))
dl.add(Dense(5, kernel_initializer="normal", activation="relu"))
dl.add(Dense(1 , kernel_initializer="normal", activation="sigmoid"))
dl.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
dl.fit(X_train,y_train, epochs=50)
```

```python
accuracy = dl.evaluate(X, y)
print('Accuracy:' % (accuracy*100))
#Değerlendirme(evaluate) işlevi, iki değer içeren bir liste döndürür. 
#Birincisi, veri kümesindeki modelin kaybı, ikincisi ise veri kümesindeki modelin doğruluğu olacaktır. 
```
